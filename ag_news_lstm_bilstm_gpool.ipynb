```python
import pandas as pd
```

```python
df=pd.read_csv('ag_news.csv')
```

```python
df
```

```python
df['Title'].isnull().sum()
```

```python
df['Class Index']
```

```python
df=df.rename(columns={'Class Index':'class'})
```

```python
df
```

```python
neworder=['Title','Description','class']
```

```python
df=df.reindex(columns=neworder)
```

```python
df
```

```python
import re
from nltk.stem import WordNetLemmatizer
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z0-9]',' ',df['Title'][i])
    review=review.lower()
    review=review.split()
    
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
sentence_length=max(len(i.split()) for i in corpus)
```

```python
sentence_length
```

```python
from tensorflow.keras.layers import TextVectorization
```

```python
tv=TextVectorization(output_mode='int',output_sequence_length=sentence_length)
```

```python
tv.adapt(corpus)
```

```python
tokenized_corpus=tv(corpus)
```

```python
corpus[4]
```

```python
tokenized_corpus[4]
```

```python
from tensorflow.keras.layers import Embedding, Dense, LSTM,Input
from tensorflow.keras.models import Sequential
```

```python
vector_representation_size=100
```

```python
model=Sequential([
    Input(shape=(sentence_length,))
])
```

```python
vocab_size=len(tv.get_vocabulary())
```

```python
print(vocab_size)
```

```python
model.add(Embedding(input_dim=vocab_size,output_dim=vector_representation_size))
```

```python
model.add(LSTM(200))
```

```python
model.add(Dense(4,activation='softmax'))
```

```python
from tensorflow.keras.optimizers import AdamW
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer=AdamW(learning_rate=1e-3),metrics=['accuracy'])
```

```python
model.summary()
```

```python
x=df.drop(labels='class',axis=1)
```

```python
y=df['class']-1 #sparse_cat_cross wants labels from 0 not from 1
```

```python
y.value_counts()
```

```python
import numpy as np
```

```python
x=np.array(tokenized_corpus)
```

```python
y=np.array(y)
```

```python
x
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
```

```python
import tensorflow 
```

```python
early_stopping=tensorflow.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    verbose=1,
    min_delta=0.001,
    patience=3,
    restore_best_weights=True,
    mode='auto'
)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=128,callbacks=[early_stopping])  
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred_labels=np.argmax(y_pred,axis=1)
```

```python
y_pred_labels
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,log_loss,roc_auc_score
```

```python
print(classification_report(y_test,y_pred_labels))
```

```python
print(confusion_matrix(y_test,y_pred_labels))
```

```python
print(accuracy_score(y_test,y_pred_labels))
```

```python
print(log_loss(y_test,y_pred))
```

```python
print(roc_auc_score(y_test,y_pred,multi_class='ovr'))
```
