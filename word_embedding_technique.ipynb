```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.preprocessing.text import one_hot #this is for one hot encoding using tf
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
vocabulary_size=10000 #this is like defining the length of your OHE matrix, it's basically the dimension, you can take any value as this is a hyperparameter
```

```python
onehot_repr=[one_hot((words),(vocabulary_size))for words in sent] #this captures the indices of where the word is present in in the matrix
```

```python
print(onehot_repr)
```

```python
from tensorflow.keras.utils import pad_sequences #we add padding as while training our NN, each of the sentence should be of the same size, in our case some are 4 and some are 5, so by applying padding (esentially just adding zeroes before or after the indice values we can make every sentence's length the same)
```

```python
sent_length=7 #i set it to 7 so that all the sentences lengrh will be 7, just keepin it constant
```

```python
embedded_docs=pad_sequences(onehot_repr,maxlen=sent_length,dtype='int32',padding='pre') #(your ohe matrix,sent_length to pad to, padding do you want it before or after)
```

```python
embedded_docs=np.expand_dims(embedded_docs,axis=0) #adding the batch dimension
```

```python
print(embedded_docs)
```

```python
dimension=10 #this is basically how many vectors we show
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Embedding,Input
```

```python
model = Sequential([
    Input(shape=(sent_length,)),   
    Embedding(input_dim=vocabulary_size,
              output_dim=dimension)
])
```

```python
model.compile('adam','mse')
```

```python
print(embedded_docs.shape)
```

```python
model.summary()
```
