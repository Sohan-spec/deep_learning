```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
```

```python
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

```python
def tokenize_sentences(text):
    return nltk.sent_tokenize(text)

def stem_text(text):
    tokens=nltk.word_tokenize(text)
    return " ".join(stemmer.stem(token) for token in tokens)

def lemmatize_text(text):
    tokens=nltk.word_tokenize(text)
    return " ".join(lemmatizer.lemmatize(token) for token in tokens)
```

```python
df['Sentences']=df[' Cluster Label'].apply(tokenize_sentences)
df['Sentences_product']=df['Product Title'].apply(tokenize_sentences)
df['Stemmed']=df[' Cluster Label'].apply(stem_text)
df['Stemmed_product']=df['Product Title'].apply(stem_text)
```

```python
df['lemmatized']=df[' Cluster Label'].apply(lemmatize_text)
df['lemmatized_product']=df['Product Title'].apply(lemmatize_text)
```
