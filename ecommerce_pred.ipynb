```python
df=df.reset_index(drop=True)
```

```python
import re
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
stop_words=stopwords.words('english')
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['desc'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stop_words]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
lengths=[len(i.split())for i in corpus]
sentence_length=int(np.percentile(lengths,99))
```

```python
print(sentence_length)
```

```python
y=df['Household']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
y=label_encoder.fit_transform(y)
```

```python
y
```

```python
y=np.array(y)
```

```python
y
```

```python
each_vector_length=100
```

```python
from tensorflow.keras.layers import Dense,LSTM,Bidirectional,Dropout,Input,Embedding
from tensorflow.keras.models import Sequential
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train_text,x_test_text,y_train,y_test=train_test_split(corpus,y,test_size=0.3,stratify=y,random_state=42)
```

```python
from tensorflow.keras.layers import TextVectorization
```

```python
tv=TextVectorization(output_mode='int',output_sequence_length=sentence_length,max_tokens=20000)
```

```python
tv.adapt(x_train_text)
x_train=tv(x_train_text)
x_test=tv(x_test_text)
```

```python
x_train.shape,y_train.shape
```

```python
import tensorflow 
```

```python
early_stopping=tensorflow.keras.callbacks.EarlyStopping(
    min_delta=0.001,
    patience=3,
    monitor='val_loss',
    restore_best_weights=True,
    verbose=1
)
```

```python
vocabulary_size=len(tv.get_vocabulary())
```

```python
vocabulary_size
```

```python
model_lstm=Sequential([
    Input(shape=(sentence_length,))
])
```

```python
model_lstm.add(Embedding(input_dim=vocabulary_size,output_dim=each_vector_length,mask_zero=True)) #mask_zero ignores the padding added
```

```python
model_lstm.add(Dropout(0.3))
```

```python
model_lstm.add(LSTM(64))
```

```python
model_lstm.add(Dropout(0.3))
```

```python
model_lstm.add(Dense(len(np.unique(y)),activation='softmax'))
```

```python
from tensorflow.keras.optimizers import AdamW
```

```python
model_lstm.compile(loss='sparse_categorical_crossentropy',optimizer=AdamW(learning_rate=1e-3,clipnorm=1.0),metrics=['accuracy'])
#the global gradient vectoor will be normalized to no more than 1
```

```python
model_lstm.summary()
```

```python
# model_lstm.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64,callbacks=[early_stopping])
```

```python
y_pred=model_lstm.predict(x_test)
```

```python
y_pred_labels=np.argmax(y_pred,axis=1)
```

```python
y_pred_labels
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,log_loss,roc_auc_score
```

```python
print(classification_report(y_test,y_pred_labels))
```

```python
def printallmetrics(name,y_true,y_pred,y_prob=None):
    print(f"{name}\n")
    print(f"accuracy:\n{accuracy_score(y_true,y_pred)}")
    print(f"classification report:\n{classification_report(y_true,y_pred)}")
    print(f"confusion matrix:\n{confusion_matrix(y_true,y_pred)}")
    
    if y_prob is not None:
        print(f"log loss:\n{log_loss(y_true,y_prob)}")
        print(f"roc auc score:\n{roc_auc_score(y_true,y_prob,multi_class='ovr')}")
```

```python
def predict_text(text):
    review=re.sub('[^a-zA-Z]',' ',text)
    review=review.lower().split()
    review=[lemmatizer.lemmatize(word) for word in review if word not in stop_words]
    review=' '.join(review)
    
    vector=tfidf.transform([review])
    
    pred_num=model_logr.predict(vector)
    pred_label=label_encoder.inverse_transform(pred_num)
    
    return pred_label[0],int(pred_num[0])
```
