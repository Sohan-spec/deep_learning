```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
```

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
df=pd.DataFrame(data=breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
df.head()
```

```python
df['label']=breast_cancer_dataset.target
```

```python
df.info()
```

```python
df['label']=df['label'].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['label'].value_counts()
```

```python
df.groupby('label').mean()
```

```python
x=df.drop(labels='label',axis=1)
```

```python
y=df['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
import tensorflow 
```

```python
tensorflow.random.set_seed(3)
```

```python
from tensorflow import keras
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
model=keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(units=20,activation='relu'),
    keras.layers.Dense(units=30,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
early_stopping=tensorflow.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    min_delta=0.0001,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=True
)
```

```python
model_history=model.fit(x_train,y_train,validation_split=0.1,epochs=50,callbacks=[early_stopping])
```

```python
model_history.history.keys()
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','test'],loc='lower right')
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper right')
plt.show()
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
```

```python
print(accuracy,loss)
```

```python
y_pred=(model.predict(x_test))
y_pred=(y_pred>=0.5).astype(int).ravel()
```
